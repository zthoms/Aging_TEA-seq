```r
library(ggplot2)
library(dplyr)
library(Seurat)
library(tidyverse)
```

# Fig 2g

```r
tf_bind <- readRDS(file = '../../02_Gating_Subsets/ped_sen_motifs_cisbp.rds')
```

```r
filtered_tf_bind <- filter(tf_bind, TF != 'ENSG00000250096_734')
```

```r
options(repr.plot.width = 12, repr.plot.height = 5)
options(repr.plot.width = 12, repr.plot.height = 5)
filtered_tf_bind %>%
  arrange(mlog10Padj) %>%
  mutate(id = factor(id, levels=c('CD8 Naive','CD8 CM','CD8 EM1','CD8 EM2','CD8 TEMRA',
                                  'CD4 Naive','CD4 CM','CD4 EM1','CD4 EM2','CD4 TEMRA','Treg'))) %>%
ggplot(aes(x = id, y = reorder(TF,mlog10Padj), color = mlog10Padj, size = mlog10Padj)) + geom_point() + scale_color_gradientn(colours = c('#006d2c','white','#d95f02'), 
                                                                                             values = scales::rescale(c(-30, -15, 0, 50, 200))) +  
theme_bw() + 
theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1, size = 15),
      axis.text.y = element_text(size = 15),
      legend.text = element_text(size = 15),
      axis.title = element_blank()) + coord_flip()
#'#1b9e77','white','#d95f02'
```
